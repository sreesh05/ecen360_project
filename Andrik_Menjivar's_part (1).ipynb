```python
category_2024 = pd.DataFrame(columns= ['TOR_F_SCALE' , 'STATE'] )

category_2024['TOR_F_SCALE'] = (data_2024['TOR_F_SCALE']).dropna().copy()#copy to
category_2024['STATE'] = (data_2024['STATE']).dropna().copy()
# category_2001_TX = category_2001[category_2001['ST']]


category_2024_TX = category_2024[(category_2024['STATE'] =='TEXAS') & (category_2024['TOR_F_SCALE']!= 'EFU')]#filter only the state of texas




```

```python
import matplotlib.pyplot as plt


state_freq = category_2024_TX['TOR_F_SCALE'].value_counts()
plt.bar( state_freq.index ,state_freq.values )
print(state_freq.index ,state_freq.values)
plt.title("Storm freq in Texas(2024)")
plt.xlabel("Category")
plt.ylabel("Freq")
plt.show()
```

```python
import pandas as pd


#graphical
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score , adjusted_rand_score

#compute the model performance
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay , confusion_matrix, precision_score , recall_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from scipy.stats import randint


# Create a list of dataframes
#Concatenate all dataframes into a single frame
dfs = [
    data_2001, data_2002, data_2003, data_2004, data_2005, data_2006, data_2007,
    data_2008, data_2009, data_2010, data_2011, data_2012, data_2013, data_2014,
    data_2015, data_2016, data_2017, data_2018, data_2019, data_2020, data_2021,
    data_2022, data_2023, data_2024
]
combined_df = pd.concat(dfs, ignore_index=True)


full_df = combined_df[combined_df['STATE'] == 'TEXAS']#filter

# Select only the important columns
texas_storms_df =full_df[['STATE', 'TOR_F_SCALE', 'DAMAGE_PROPERTY' , 'DAMAGE_CROPS' ,'INJURIES_DIRECT', 'INJURIES_INDIRECT' , 'DEATHS_DIRECT']]

# Drop rows with any null values
texas_storms_df = texas_storms_df.dropna()
texas_storms_df = texas_storms_df[texas_storms_df['TOR_F_SCALE'] != 'EFU']


#format the numbers K(thousands), M(millions) to floating points
def convert_num(val):
    if pd.isna(val) or val.strip() == '':
        return 0
    val = val.strip().upper()
    try:
        if val.endswith('K'):
            return int(val[:-1]) * 1_000 # convert k to a 1000
        elif val.endswith('M'):
            return int(val[:-1]) * 1_000_000 # convert M to 1,000,000
        elif val.endswith('B'):
            return int(val[:-1]) * 1_000_000_000
        else:
            return int(val)
    except ValueError:
        return 0  #edge case


#apply the number conversions
texas_storms_df['DAMAGE_PROPERTY'] = texas_storms_df['DAMAGE_PROPERTY'].apply(convert_num)
texas_storms_df['DAMAGE_CROPS'] = texas_storms_df['DAMAGE_CROPS'].apply(convert_num)


#for supervised
thresholds = {

    1: {
        'category': ('F0','F1' , 'EF0', 'EF1'),
        'fatalities': (0, 1),
        'indirect_injuries': (0, 30),
        'direct_injuries': (0, 20),
        'property_damage': (0, 10000),  # Range in dollars
        'crop_damage': (0, 5000)      # Range in dollars
    },
    2: {
        'category': ( 'F2', 'F3' ,'EF2', 'EF3'),
        'fatalities': (3, 8),
        'indirect_injuries': (5, 30),
        'direct_injuries': (10, 50),
        'property_damage': (10000, 100000),  # Range in dollars
        'crop_damage': (5000, 20000)      # Range in dollars
    },
    3: {
        'category': ('F4' , 'F5' , 'EF4', 'EF5'),
        'fatalities': (4, 15),
        'indirect_injuries': (20, 60),
        'direct_injuries': (30, 100),
        'property_damage': (100000, 1000000),  # Range in dollars
        'crop_damage': (20000, 100000)      # Range in dollars
    }
}
#--------------------------------#


#label the severity categories for each entry
def label_severity(dat, thresholds):
    best_classify = None
    best_score = float('inf')  #score

    # Loop through all categories to check for a match
    for classify, ranges in thresholds.items():
        # Check if the storm's F-Scale is within the defined category
        if dat['TOR_F_SCALE'] in ranges['category']:
            # Calculate "distance score" for this category
            score_dist = 0

            # Calculate distances for each feature
            score_dist += max(0, dat['DEATHS_DIRECT'] - ranges['fatalities'][1])  # Above max fatalities
            score_dist += max(0, ranges['fatalities'][0] - dat['DEATHS_DIRECT'])  # Below min fatalities

            score_dist += max(0, dat['INJURIES_DIRECT'] - ranges['direct_injuries'][1])  # Above max injuries
            score_dist += max(0, ranges['direct_injuries'][0] - dat['INJURIES_DIRECT'])  # Below min injuries

            score_dist += max(0, dat['INJURIES_INDIRECT'] - ranges['indirect_injuries'][1])  # Above max indirect injuries
            score_dist += max(0, ranges['indirect_injuries'][0] - dat['INJURIES_INDIRECT'])  # Below min indirect injuries

            score_dist += max(0, dat['DAMAGE_PROPERTY'] - ranges['property_damage'][1])  # Above max property damage
            score_dist += max(0, ranges['property_damage'][0] - dat['DAMAGE_PROPERTY'])  # Below min property damage

            score_dist += max(0, dat['DAMAGE_CROPS'] - ranges['crop_damage'][1])  # Above max crop damage
            score_dist += max(0, ranges['crop_damage'][0] - dat['DAMAGE_CROPS'])  # Below min crop damage

            # If this score is the best (smallest) so far, update the best classification
            if score_dist < best_score:
                best_score = score_dist
                best_classify = classify

    # Return the category with the closest match
    return best_classify if best_classify is not None else "Failed"



#map out categories based on these values
texas_storms_df['severity level'] = texas_storms_df.apply(lambda x: label_severity(x, thresholds), axis=1)

```

K-means clustering(Unsupervised)

```python

#Building the ML model
#search for best optimization hyper parameters
optimization = { 'max_depth': randint(1, 6),
    'n_estimators': randint(10, 200)}

#training columns and testing columns
X_dt = texas_storms_df.drop([ 'STATE' , 'TOR_F_SCALE', 'severity level'], axis=1)
y_dt = texas_storms_df['severity level']


#implement K-means clustering for groups of storm severities
clusters_k = 3
kmeans = KMeans(n_clusters=clusters_k, random_state=42)
kmeans.fit_predict(X_dt)


#map clusters to severity levels for prediction
severity_mapping = {
    0: 1,  # Cluster 0 corresponds to severity level 1
    1: 2,  # Cluster 1 corresponds to severity level 2
    2: 3   # Cluster 2 corresponds to severity level 3
}

#apply label
y_pred_mapped = [severity_mapping[label] for label in kmeans.labels_]

#mapped labels
y_pred_mapped





```

```python
sil_score = silhouette_score(X_dt, kmeans.labels_)
print("Silhouette Score:", sil_score)
```

```python
#output prediction and actual
print("Predicted severity levels:")
print(y_pred_mapped[50:100])
print("\nActual severity levels:")
print(y_dt[50:100])
```

```python
#confusion matrix

confu = confusion_matrix(y_dt, y_pred_mapped)

ConfusionMatrixDisplay(confu).plot()
plt.show()
```

Further testing Accuracy of model

```python
accuracy = accuracy_score(y_dt, y_pred_mapped)
precision = precision_score(y_dt, y_pred_mapped, average='weighted')
recall = recall_score(y_dt, y_pred_mapped, average='weighted')
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

```python
from sklearn.metrics import adjusted_rand_score

ari = adjusted_rand_score(y_dt, y_pred_mapped)
print(f"Adjusted Rand Index: {ari}")
```

RandomForestModel(Supervised)

```python
# #SPLIT THE training and testing set(X,y)
X_test, X_train, y_test, y_train = train_test_split(X_dt , y_dt ,  test_size=0.2, random_state=42)

#make the model
model_Ml = RandomForestClassifier(n_estimators=100 , random_state=42)

search =RandomizedSearchCV(model_Ml , param_distributions=optimization , n_iter =3 , cv=5 , random_state=42)
search.fit(X_train, y_train)


#prediction output
y_pred = search.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
#Confusion matrix
confu = confusion_matrix(y_test, y_pred)

ConfusionMatrixDisplay(confu).plot()
plt.show()
```
